```python
import numpy as np
import matplotlib.pyplot as plt
# Definimos un rango de valores candidatos para la carga elemental
candidatos = np.linspace(0, max(q),100000)

errores_promedio = []

# Recorremos todos los candidatos
for q_i in candidatos:
  if q_i!=0:
    # Dividimos cada carga por el candidato
    cocientes = cargas / q_i
    # Redondeamos los cocientes al entero mÃ¡s cercano
    redondeados = np.round(cocientes)
    # Calculamos el error promedio
    error = np.mean(np.abs(cocientes - redondeados))
    errores_promedio.append(error)

# Encontramos el candidato con menor error
mejor_indice = np.argmin(errores_promedio)
mejor_carga = candidatos[mejor_indice]

print(f"La mejor estimaciÃ³n de la carga fundamental es: {mejor_carga:.3e} C")
```
